```python
import numpy as np
import pandas as pd
import pygimli as pg
from pygimli.physics import ert
import matplotlib.pyplot as plt
import glob
import os

```

```python
gil = pd.read_csv('../climate data/en_1991-2020_Normals_MB_GILLAM.csv')
chu = pd.read_csv('../climate data/en_1991-2020_Normals_MB_CHURCHILL.csv')
```

```python
MAAT_gil = gil[
    (gil["PERIOD_OF_RECORD"] == "Normal") &
    (gil["ELEMENT_GROUP"] == "Temperature") & 
    (gil['NORMALS_ELEMENT']== 'Daily Average (°C)')
]['Year']

MAAT_chu = chu[
    (chu["PERIOD_OF_RECORD"] == "Normal") &
    (chu["ELEMENT_GROUP"] == "Temperature") & 
    (chu['NORMALS_ELEMENT']== 'Daily Average (°C)')
]['Year']

```

```python
print(MAAT_gil)
print('')
print(MAAT_chu)
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

precip_chu = np.array(chu[
    (chu["PERIOD_OF_RECORD"] == "Normal") &
    (chu["ELEMENT_GROUP"] == "Precipitation") & 
    (chu['NORMALS_ELEMENT']== 'Precipitation (mm)')
][months].iloc[0]).astype(float)

precip_gil = np.array(gil[
    (gil["PERIOD_OF_RECORD"] == "Normal") &
    (gil["ELEMENT_GROUP"] == "Precipitation") & 
    (gil['NORMALS_ELEMENT']== 'Precipitation (mm)')
][months].iloc[0]).astype(float)


temp_chu = np.array(chu[
    (chu["PERIOD_OF_RECORD"] == "Normal") &
    (chu["ELEMENT_GROUP"] == "Temperature") & 
    (chu['NORMALS_ELEMENT']== 'Daily Average (°C)')
][months].iloc[0]).astype(float)

temp_gil = np.array(gil[
    (gil["PERIOD_OF_RECORD"] == "Normal") &
    (gil["ELEMENT_GROUP"] == "Temperature") & 
    (gil['NORMALS_ELEMENT']== 'Daily Average (°C)')
][months].iloc[0]).astype(float)
```

```python
temp_chu
```

```python
fs = 16
c1 = 'dodgerblue'
c2 = 'orange'

fig, ax1 = plt.subplots(figsize=[10,6])

# plt.bar(months,precip_chu)
# plt.bar(months,precip_gil)

x = np.arange(len(months))
width = 0.3  

ax1.plot(x, temp_chu,color=c1,label='Churchill temperature')
ax1.plot(x, temp_gil,color=c2,label='Gillam temperature')

ax2 = ax1.twinx()

rects1 = ax2.bar(x - width/2, precip_chu, width, label='Churchill precipitation',color=c1,alpha=0.5)
rects2 = ax2.bar(x + width/2, precip_gil, width, label='Gillam precipitation',color=c2,alpha=0.5)

ax1.set_ylabel(r'Temperature ($\degree$C)', fontsize=fs)
ax2.set_ylabel('Precipitation (mm)',fontsize=fs,labelpad=10)

ax1.set_xticks(x)
ax1.set_xticklabels(months, fontsize=fs-2)  # optional rotation

ax1.tick_params(axis='both', labelsize=fs-2)
ax2.tick_params(axis='both', labelsize=fs-2)

ax1.set_zorder(3)
ax2.set_zorder(2)
ax1.patch.set_visible(False)

handles1, labels1 = ax1.get_legend_handles_labels()
handles2, labels2 = ax2.get_legend_handles_labels()
handles = handles1 + handles2
labels = labels1 + labels2

plt.legend(handles, labels,fontsize=fs-2)

ax1.set_ylim([-30,20])
ax2.set_ylim([0,100])

plt.grid()
ax2.set_axisbelow(True)
```
